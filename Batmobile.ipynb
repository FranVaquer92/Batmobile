Then, this is the code:


```
class batmobile:
  def __init__(self, pilot, n_batrockets, n_pem, gas):
    if pilot == 'Batman' or pilot == 'Robin':
      self.pilot = pilot
    else:
      raise Exception("You are not worthy to drive the batmobile")
    if isinstance(n_batrockets, rocket):
      self.n_batrockets = n_batrockets
    else:
      raise Exception("There aren't rockets!")
    if isinstance(n_pem, pem):
      self.n_pem = n_pem
    else:
      raise Exception("There aren't pem!")
    self.gas = gas
  def describe_status(self):
    print(f"The batmobile is drived by {self.pilot}, has {self.n_batrockets.ammunition} batrockets and {self.n_pem.ammunition} pem. {self.gas} % of gas reamins")
  def accelerate(self, distance):
    if (0.5*distance) <= self.gas:
      self.gas = self.gas-(0.5*distance)
      print(f"{self.pilot} drive for {distance} km and {self.gas} % of gas remains")
    else:
      raise Exception("There are not gas! You can't accelerate")
  def turbo(self, distance):
    if (1*distance) <= self.gas:
      self.gas = self.gas-(1*distance)
      print(f"{self.pilot} drive fastly for {distance} km and {self.gas} % of gas remains")
    else:
      raise Exception("There are not gas! You can't use the turbo")
  def launch_batrocket(self, enemy):
    if self.n_batrockets.ammunition > 0:
      self.n_batrockets.consum_ammunition()
      print(f"There are {self.n_batrockets.ammunition} batrockets")
      enemy.joker_ko()
    else:
      raise Exception("There aren't batrockets, you must recharge")
  def launch_pem(self, enemy):
    if self.n_pem.ammunition > 0:
      self.n_pem.consum_ammunition()
      print(f"There are {self.n_pem.ammunition} pem")
      enemy.joker_unweapon()
    else:
      raise Exception("There aren't pem, you must recharge")

class joker:
  def __init__(self,KO=False, weapon=True):
    self.KO = KO
    self.weapon = weapon
  def joker_ko(self):
    self.KO = True
    print(f"The joker is dead")
  def joker_unweapon(self):
    self.weapon = False
    print(f"The joker is unarmed")
  def describe_status(self):
    if self.KO == True:
      print(f"The joker is dead")
    elif self.weapon == True:
      print(f"Oh my god! The joker is alive and armed!")
    else:
      print(f"The joker is alive but he is unarmed")
      
  
class joker_villain:
  def __init__(self,KO=False, weapon=True):
    self.KO = KO
    self.weapon = weapon
  def joker_ko(self):
    self.KO = True
    print(f"The joker villain is dead")
  def joker_unweapon(self):
    self.weapon = False
    print(f"The joker villain is unarmed")
  def describe_status(self):
    if self.KO == True:
      print(f"The joker is dead")
    elif self.weapon == True:
      print(f"Oh my god! The joker villain is alive and armed!")
    else:
      print(f"The joker villain is alive but he is unarmed")

class rocket:
  def __init__(self, ammunition):
    self.ammunition = ammunition
  def recharge_ammunition(self, recharge):
    self.ammunition = self.ammunition + recharge
    print(f"Good job! The rocket arsenal is recharged to {self.ammunition}")
  def consum_ammunition(self):
    self.ammunition = self.ammunition - 1

class pem:
  def __init__(self, ammunition):
    self.ammunition = ammunition
  def recharge_ammunition(self, recharge):
    self.ammunition = self.ammunition + recharge
    print(f"Good job! The pem arsenal is recharged to {self.ammunition}")
  def consum_ammunition(self):
    self.ammunition = self.ammunition - 1
```
